# Data Cleaning and Pre-Processing:

### Check for NULL values:

```python
# Check for Null values
print('Check for Null values')
df.isnull().sum()
```

```python
df.isnull().any()
```

### Null values found for age, location, size and review rating

```python
# Replace NULL values for Age:
print('Replace NULL values for Age:')
print(f'Median for age:{df['Age'].median()}')
print(f'Replace NULL age with: {df['Age'].median()}')
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# Replace NULL values for Location:
print('Replace NULL values for Location:')
print(f'Mode for Location:{df['Location'].mode()[0]}')
print(f'Replace NULL Location with: {df['Location'].mode()[0]}')
df['Location'] = df['Location'].fillna(df['Location'].mode()[0])
```

```python
# Replace NULL values for Size:
print('Replace NULL values for Size:')
print(f'Mode for Size:{df['Size'].mode()[0]}')
print(f'Replace NULL Size with: {df['Size'].mode()[0]}')
df['Size'] = df['Size'].fillna(df['Size'].mode()[0])
```

```python
# Replace NULL values for Review Rating:
print('Replace NULL values for Review Rating:')
print(f'Mean for Review Rating:{df['Review Rating'].mean()}')
print(f'Replace NULL Review Rating with: {df['Review Rating'].mean()}')
df['Review Rating'] = df['Review Rating'].fillna(df['Review Rating'].mean())
```

# Exploratory Data Analysis (EDA):

## Univariate Analysis:

```python
# Check columns:
df.columns
```

```python
# Gender distrinution
df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
print('Majority Male customers are found!')
```

```python
# Category distrinution
df['Category'].value_counts().plot(kind='pie')
plt.title("Product Category Distribution")
plt.show()
print('Clothing category is most purchased!')
```

```python
# Age Distribution
sns.histplot(df['Age'], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
print('40-50 years old customers purchased the most!')
```

```python
# Payment Method Distribution
df['Payment Method'].value_counts().plot(kind='pie')
plt.title("Payment Method Distribution")
plt.show()
print('Cash, Credit Card and Paypal was most used payment methods!')
```

## Bivariate Analysis:

```python
# Category-wise Sales:
df.groupby('Category')['Purchase Amount (USD)'].sum().plot(kind='barh')
plt.title("Total Sales by Category")
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
print('Clothing category earned most revenue!')
```

```python
# Sales By Season:
df.groupby('Season')['Purchase Amount (USD)'].sum().plot(kind='line')
plt.title("Sales by Season")
plt.xlabel('Season')
plt.ylabel('Total Sales')
plt.show()
print('Fall season is the best season for shopping!')
```

```python
# Total Spending By Subscription Status:
df.groupby('Subscription Status')['Purchase Amount (USD)'].sum().plot(kind='pie')
plt.title("Total Spending by Subscription Status")
plt.show()
print('Subscribed Customers does not generate more revenue!')
```

```python
# Discount impact on Purchase Amount:
sns.barplot(x='Discount Applied',
            y='Purchase Amount (USD)',
            data=df)
plt.title("Discount Impact on Purchase Amount")
plt.show()
print('Discount does not increase spending!')
```

```python
# Gender vs Payment Method:
pd.crosstab(df['Gender'], df['Payment Method']).plot(kind='bar')
plt.title("Gender vs Payment Method")
plt.show()
print('Most Male customers used Cash payment and Female customers used Credit Card Payment options!')
```

```python
# Top 10 locations by Total Sales:
top10locsales = df.groupby('Location')['Purchase Amount (USD)'].sum()
top10locsales = top10locsales.sort_values(ascending=False).head(10)
top10locsales.plot(kind='bar')
plt.title("Top 10 Locations by Total Sales")
plt.xlabel("Location")
plt.ylabel("Total Sales")
plt.show()
print('Montana is the top selling city!')
```

```python
# Top 10 locations by Customers:
top10loccust = df.groupby('Location')['Customer ID'].count()
top10loccust = top10loccust.sort_values(ascending=False).head(10)
top10loccust.plot(kind='barh')
plt.title("Top 10 Locations by Customers")
plt.xlabel("Location")
plt.ylabel("Total Customers")
plt.show()
print('Most customers are from Montana!')
```

```python
# Top 10 locations by Reviews:
top10locrev = df.groupby('Location')['Review Rating'].mean()
top10locrev = top10locrev.sort_values(ascending=False).head(10)
top10locrev.plot(kind='pie', autopct='%1.1f%%')
plt.title("Top 10 Locations by Reviews")
plt.xlabel("Location")
plt.ylabel("Average Reviews")
plt.show()
print('Most reviews are given by customers from Texas and Wisconsin!')
```
